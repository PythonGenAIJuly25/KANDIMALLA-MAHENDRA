```python
# Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.
def trans(l):
  n_l=[[0]*3 for i in range(3)]
  for i in range(3):
    for j in range(3):
      n_l[i][j]=l[j][i]
  return n_l

l=[[1,2,3],[4,5,6],[7,8,9]]
print("Original matrix :")
for i in range(3):
  for j in range(3):
    print(l[i][j],end=" ")
  print()

trans_list=trans(l)
print("Transposed list is : ")
for i in range(3):
  for j in range(3):
    print(trans_list[i][j],end=" ")
  print()
```

```python
# Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.
def flat(l):
  nl=[]
  for i in range(3):
    for j in range(3):
      nl.append(l[i][j])
  return nl

l=[[1,2,3],[4,5,6],[7,8,9]]
print("Original matrix :")
for i in range(3):
  for j in range(3):
    print(l[i][j],end=" ")
  print()

flat_list=flat(l)
print("Falten list is : " , flat_list)
```

```python
# Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists.
def rev(l):
  return l[::-1]
l1=[1,2,3,4,5]
print("Original list : ",l1)
print("reverse list : ",rev(l1))
```

```python
# Write a function that rotates a list by n positions. Print the original and rotated lists.
def rotate(l):
  return l[3::]+l[0:3]
l=list(map(int,input("Enter list elemenet: ").split(" ")))
n=int(input("Enter positions : "))
print("original list : ",l)
print("rotated list : ",rotate(l))

```

```python
# Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.
def inter(l1,l2):
  l3=[x for x in l1 if x in l2]
  return l3
l1=[1,2,3,4,5,6]
l2=[4,5,6,7,8,9]
print("Intersected list is :",inter(l1,l2))
```
