```python
import matplotlib.pyplot as plt
```

```python
x=range(1,6)
y=[i**2 for i in x]
plt.plot(x,y)
# y=x**2

plt.xlabel("x label")
plt.ylabel("y label")
plt.title("y = x**2")
plt.show()
```

```python
x=range(1,6)
y=[i**3 for i in x]
plt.figure(figsize=(4,2))
plt.plot(x,y,color='r',linestyle='dashed',marker='h')
plt.show()
```
