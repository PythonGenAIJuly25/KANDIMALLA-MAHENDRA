```python
# Assignment 1: Define a function that takes a single integer as input and returns its square. Test the function with different inputs.
def square(n):
    return n * n
print(square(2))
print(square(5))
```

```python
# Assignment 2: Define a function that takes two integers as input and returns their sum. Test the function with different inputs.
def add(a, b):
    return a + b
print(add(3, 4))
print(add(10, 20))
```

```python
# Assignment 3: Define a function that takes two integers as input and returns their sum. The second integer should have a default value of 5. Test the function with different inputs.
def add_default(a, b=5):
    return a + b
print(add_default(3))
print(add_default(3, 7))
```

```python
# Assignment 4: Define a function that takes three named arguments: first_name, last_name, and age, and returns a formatted string. Test the function with different inputs.
def format_person(first_name, last_name, age):
    return f"{first_name} {last_name} is {age} years old"
print(format_person(first_name="John", last_name="Doe", age=30))
print(format_person(first_name="Alice", last_name="Smith", age=25))
```

```python
# Assignment 5: Define a function that takes a variable number of integer arguments and returns their product. Test the function with different inputs.
def product(*args):
    result = 1
    for num in args:
        result *= num
    return result
print(product(2, 3))
print(product(1, 2, 3, 4))
```

```python
# Assignment 6: Define a function that contains another function inside it. The outer function should take two integers as input and return the result of the inner function, which multiplies the two integers. Test the function with different inputs.
def outer(a, b):
    def inner(x, y):
        return x * y
    return inner(a, b)
print(outer(2, 3))
print(outer(4, 5))
```

```python
# Assignment 7: Define a function that takes a single integer as input and returns the integer squared, cubed, and raised to the power of four. Test the function with different inputs.
def powers(n):
    return n**2, n**3, n**4
print(powers(2))
print(powers(3))
```

```python
# Assignment 8: Define a recursive function that calculates the factorial of a given number. Test the function with different inputs.
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
print(factorial(5))
print(factorial(3))
```

```python
# Assignment 9: Define a lambda function that takes two integers as input and returns their sum. Test the lambda function with different inputs.
sum_lambda = lambda a, b: a + b
print(sum_lambda(2, 3))
print(sum_lambda(5, 10))
```

```python
# Assignment 10: Use the map function to apply a lambda function that squares each number in a list of integers. Test with different lists.
print(list(map(lambda x: x**2, [1, 2, 3, 4])))
print(list(map(lambda x: x**2, [5, 6, 7])))
```

```python
# Assignment 11: Use the filter function to filter out all odd numbers from a list of integers. Test with different lists.
print(list(filter(lambda x: x % 2 == 0, [1, 2, 3, 4, 5])))
print(list(filter(lambda x: x % 2 == 0, [10, 15, 20, 25])))
```

```python
# Assignment 12: Define a decorator function that prints 'Executing function...' before executing a function and 'Function executed.' after executing it. Apply this decorator to a function that takes a list of integers and returns their sum. Test the decorated function with different lists.
def decorator(func):
    def wrapper(*args, **kwargs):
        print("Executing function...")
        result = func(*args, **kwargs)
        print("Function executed.")
        return result
    return wrapper
@decorator
def sum_list(numbers):
    return sum(numbers)

print(sum_list([1, 2, 3]))
print(sum_list([10, 20, 30]))
```

```python
# Assignment 13: Define a function that takes variable-length arguments and keyword arguments and prints them. Test the function with different inputs.
def print_args_kwargs(*args, **kwargs):
    print(args)
    print(kwargs)
print_args_kwargs(1, 2, 3, name="Alice", age=25)
print_args_kwargs("a", "b", x=10, y=20)
```

```python
# Assignment 14: Define a higher-order function that takes a function and a list of integers as arguments, and applies the function to each integer in the list. Test with different functions and lists.
def apply_func(func, nums):
    return [func(n) for n in nums]
print(apply_func(lambda x: x + 1, [1, 2, 3]))
print(apply_func(lambda x: x * 2, [4, 5, 6]))
```

```python
# Assignment 15: Define a function with a docstring that explains what the function does, its parameters, and its return value. Print the function's docstring.
def multiply(a, b):
    """
    Multiplies two numbers.

    Parameters:
    a (int): First number
    b (int): Second number

    Returns:
    int: Product of a and b
    """
    return a * b

print(multiply.__doc__)
```
